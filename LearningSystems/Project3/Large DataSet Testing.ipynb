```python
data = pd.read_csv('Churn_Modelling.csv')

data.head(4)
data.info()
data.describe()
data.tail()
data.isnull().sum()
```

```python
data = data.drop(['CustomerId', 'Surname', 'RowNumber'], axis = 1)



print(data.columns)
data.shape
```

```python
x = data.iloc[:,0:10]
y = data.iloc[:,10]

print(x.shape)
print(y.shape)

print(x.columns)
```

```python
x = pd.get_dummies(x)

x.head()
```

```python

x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)

x_train = pd.DataFrame(x_train)
x_train.head()
```

```python
#Decision Tree Induction
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix

model = DecisionTreeClassifier() 
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

print("The Training Accuracy of Decision Tree Model :", model.score(x_train, y_train))
print("The Testing Accuaracy of Decision Tree Model :", model.score(x_test, y_test))

cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
#Calculating Memory used
process = psutil.Process(os.getpid())
print("Total Used Memory for Decision Tree: ", end='')
print(process.memory_info().rss)
```

```python
#SVM Model

from sklearn.svm import SVC

model = SVC()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

print("The Training Accuracy Of SVM Model :", model.score(x_train, y_train))
print("The Testing Accuracy of SVM Model :", model.score(x_test, y_test))

cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
process = psutil.Process(os.getpid())
print("Total Used Memory for SVM: ", end='')
print(process.memory_info().rss)
```

```python
#ANN
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# creating the model
model = Sequential()

# first hidden layer
model.add(Dense(units=6, activation = 'relu')) #input/first hidden layer
model.add(Dense(units=6, activation = 'relu')) #second hidden layer
model.add(Dense(units=1, activation = 'sigmoid')) #output layer

# Compiling the NN

model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy']) 

model.fit(x_train, y_train, batch_size = 10, epochs = 50)
```

```python
y_pred = model.predict(x_test)
y_pred = (y_pred > 0.5)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
print("The accuracy of ANN",(accuracy_score(y_test,y_pred))*100,"%")

```

```python
process = psutil.Process(os.getpid())
print("Total Used Memory for ANN: ", end='')
print(process.memory_info().rss)
```
