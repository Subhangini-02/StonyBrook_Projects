```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, accuracy_score,precision_score,recall_score
from sklearn.metrics import classification_report
```

```python
dataset = pd.read_csv('Indian Liver Patient Dataset (ILPD).csv')
```

```python
dataset
```

```python
Split=dataset.values
X = Split[:,2:4]
y = Split[:, 4]
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.5, random_state = 0)
```

```python

sc = StandardScaler()
# print(sc)
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
# print(X_test)
# print(X_train)
```

```python
y_train
```

```python

model=SVC(kernel='linear', C= 1)
model.fit(X_train,y_train)
```

```python
y_predict = model.predict(X_test)
print("Accuracy of the data set is: %.2f" % accuracy_score(y_test,y_predict))
print("Data Sensitivity: %.2f" % precision_score(y_test,y_predict))
print("Data Specificity: %.2f" % recall_score(y_test,y_predict))
```
